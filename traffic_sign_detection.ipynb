```python
import cv2
from ultralytics import YOLO
```

```python
model = YOLO('traffic_sign_detector.pt')  # Replace with the actual path to your model
```

```python
cap = cv2.VideoCapture(0)  # 0 usually refers to the default webcam
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)  # Perform inference

    # Visualize results on the frame (e.g., using results.plot())
    annotated_frame = results[0].plot()  

    cv2.imshow('YOLOv11 Webcam', annotated_frame)

    # --- MODIFIED EXIT FEATURE ---
    # cv2.waitKey(1) returns the ASCII code of the pressed key (or -1 if none).
    key = cv2.waitKey(1) & 0xFF 
    
    # 1. Exit on 'q' press (Original logic)
    if key == ord('q'):
        break
    
    # 2. FEATURE ADDED: Exit on 'ESC' press (Common practice)
    if key == 27: # 27 is the ASCII code for the ESC key
        break
        
    # 3. ALTERNATIVE: Exit on ANY key press (Replace 1 and 2 with this)
    # if cv2.waitKey(1) != -1:
    #     break

# Release the video capture object and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
